```python

import os

nome = input("\nQual seu nome? ")
altura = float(input("\nQual a sua altura?"))
peso = int(input("\nQual o seu peso?"))
os.system('clear')
print(nome)
print(altura, 'm')
print(peso, 'kg')
imc = peso / (altura**2)
if imc >= 18.5 and imc <= 24.9:
    print("peso normal")
elif imc >= 25 and imc >= 29.9:
    print("sobrepeso")
elif imc < 18.5:
    print("abaixo do peso")
```
